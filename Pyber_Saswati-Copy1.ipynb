```python
# Import Dependencies
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns

import os
if os.path.exists('README.txt'):
    long_description = open('README.txt').read()
```

```python
# Import our data into pandas from CSV
st1 = "city_data.csv"
city_df = pd.read_csv(st1)
city_df
```

```python
# Import our data into pandas from CSV
st2 = "ride_data.csv"
ride_df = pd.read_csv(st2)
ride_df
```

```python
# Merging the two files 
combined_df = pd.merge(ride_df, city_df, on='city')
combined_df.head()
```

```python
city_groups = combined_df.groupby('city')

tot_rides = city_groups['ride_id'].count().reset_index().rename(columns={0:'tot_rides'})
#tot_rides

tot_driver = city_groups['driver_count'].sum().reset_index().rename(columns={2:'tot_driver'})
#tot_driver

avg_fare = city_groups['fare'].mean().reset_index().rename(columns={1:'avg.fare'})
avg_fare
```

```python
combined_df2 = combined_df.drop(columns=['date','fare','ride_id','driver_count'])
cb3 = combined_df2.groupby(['city','type']).size().reset_index().rename(columns={0:'count'})
cb4 = cb3.drop(columns='count')
cb4
```

```python
merge1 = pd.merge(cb4, avg_fare, on='city')
merge1.head()
```

```python
merge2 = pd.merge(merge1, tot_driver, on='city')
merge2.head()
```

```python
merge3 = pd.merge(merge2, tot_rides, on='city')
merge3.head()
```

```python
conditions = [
    (merge3['type'] == 'Urban'),
    (merge3['type'] == 'Suburban'),
    (merge3['type'] == 'Rural')]
choices = [1, 2, 3]
merge3['type_color'] = np.select(conditions, choices, default=4)
merge3 = merge3.set_index('city')
merge3.head()
```

```python
colors = ["yellowgreen", "red", "blue"]
merge3.plot(kind="scatter", x="ride_id", y="fare", alpha=1.0,c=colors)

plt.title("Averge fare, No of Rides and Drivers by City ")
plt.xlabel("No of Rides by City")
plt.ylabel("Average Fare by City")

plt.show()
```

```python
colors = ["yellowgreen", "red", "blue"]
merge3.plot(kind="scatter", x="driver_count", y="fare", alpha=1.0,c=colors)

plt.title("Averge fare, No of Rides and Drivers by City ")
plt.xlabel("No of Drivers by City")
plt.ylabel("Average Fare by City")

plt.show()
```

```python
colors = ["yellowgreen", "red", "blue"]
merge3.plot(kind="scatter", x="ride_id", y="driver_count", alpha=1.0,c=colors)

plt.title("Averge fare, No of Rides and Drivers by City ")
plt.xlabel("No of Rides by City")
plt.ylabel("No of Drivers by City")

plt.show()
```

```python
#% of Total Fares by City Type
typegp = combined_df.groupby('type')
total_fare = typegp['fare'].sum().reset_index().rename(columns={0:'count'})
total_fare = total_fare.set_index('type')


total_fare2 = total_fare.plot(kind="pie", y="fare", autopct="%1.1f%%", startangle=90, title=("% of Total Fares by City Type"))
plt.show()
```

```python
#% of Total Rides by City Type
total_ride = typegp['ride_id'].count().reset_index().rename(columns={0:'count'})
total_ride = total_ride.set_index('type')
total_ride2 = total_ride.plot(kind="pie", y="ride_id", autopct="%1.1f%%", startangle=90, title=("% of Total Rides by City Type"))
plt.show()
```

```python
# % of Total Drivers by City Type

total_Driver = typegp['driver_count'].sum().reset_index().rename(columns={0:'Dsum'})
total_Driver = total_Driver.set_index('type')
total_Driver2 = total_Driver.plot(kind="pie", y="driver_count", autopct="%1.1f%%", startangle=90,title=("% of Total Drivers by City Type"))
plt.show()
```
